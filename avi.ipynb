```python
y=data.iloc[:,14].values
print(y)
```

```python
x=data.iloc[:,0:14].values
print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
```

```python
X_train_std=st.fit_transform(X_train)
X_test_std=st.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
lr=LogisticRegression()
```

```python
pipe_lr=Pipeline([('std',StandardScaler()),
                  ('pca',PCA()),
                  ('lr',LogisticRegression(penalty='l1'))])
```

```python
pipe_lr.fit(X_train_std,y_train)
acc=pipe_lr.score(X_test_std,y_test)
```

```python
score=[]
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=5,random_state=None)
for train_index, test_index in skf.split(x,y):
    #print(train_index)
    X_train_std,X_test_std=x[train_index],x[test_index]
    y_train,y_test=y[train_index],y[test_index]
    pipe_lr.fit(X_train_std,y_train)
    score1=pipe_lr.score(X_test_std,y_test)
    score.append(score1)
print(score)
print("Average\n:",np.mean(score))
```

```python
from sklearn.svm import SVC 
svc=SVC()
```

```python
svc.fit(X_train_std,y_train)
```

```python
y_pred = svc.predict(X_test_std)
```

```python
print("Accuracy is:")
svc.score(X_test_std,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train_std,y_train)
```

```python
y_pred = rfc.predict(X_test_std)
```

```python
rfc.score(X_test_std,y_test)
```

# precision

```python
from sklearn.metrics import precision_score,classification_report
from sklearn.metrics import recall_score,f1_score
```

```python
print(precision_score(y_test,y_pred))
```

```python
print(f1_score(y_test,y_pred))
```

```python
print(recall_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred)) 
```

# Confusion metrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(5,5))
ax.matshow(cnf_matrix,cmap=plt.cm.Reds,alpha=0.6)
for i in range(cnf_matrix.shape[1]):
    for j in range(cnf_matrix.shape[0]):
        ax.text(x=j,y=i,s=cnf_matrix[i,j],va='center',ha='center')
plt.show()
```
