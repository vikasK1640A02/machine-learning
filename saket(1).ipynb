# Pipe line

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
lr=LogisticRegression()
```

```python
pipe_lr=Pipeline([('std',StandardScaler()),
                  ('pca',PCA()),
                  ('lr',LogisticRegression(penalty='l1'))])
```

```python
pipe_lr.fit(X_train_std,y_train)
acc=pipe_lr.score(X_test_std,y_test)
```

```python
print("accuracy using pipe-line is:",acc)
```

# Hyper tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
pipe_svc=Pipeline([('srd',StandardScaler()),('clf',SVC())])
param={'clf__C':[0.0001,0.001,0.01,0.1,1,10,100,1000,10000],'clf__gamma':[0.0001,0.001,0.01,0.1,1,10,100,1000,10000],'clf__kernel':['linear','rbf']}
```

```python
gs=GridSearchCV(estimator=pipe_svc,param_grid=param,scoring='accuracy',cv=10)
```

```python
gs.fit(X_train_std,y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train_std,y_train)
```

```python
y_pred = rfc.predict(X_test_std)
```

```python
print("Accuracy is:")
rfc.score(X_test_std,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(5,5))
ax.matshow(cnf_matrix,cmap=plt.cm.Reds,alpha=0.6)
for i in range(cnf_matrix.shape[1]):
    for j in range(cnf_matrix.shape[0]):
        ax.text(x=j,y=i,s=cnf_matrix[i,j],va='center',ha='center')
plt.show()
```
